## group by()

```python
import pandas as pd
```

```python
df.groupby(by="Loan_ID")["Self_Employed","Credit_History"].first()
```

```python
df.groupby(by="Loan_ID")["Self_Employed"].last()
```

## crosstab

```python
##syntax=  pd.crosstab(column1,column 2)
```

```python
pd.crosstab(df["Gender"],df["Married"])
```

```python
pd.crosstab(df["Gender"],df["Education"])
```

```python
pd.crosstab(df["Gender"],df["ApplicantIncome"])
```

```python
pd.crosstab(df["Gender"],df["Married"],values=df["ApplicantIncome"],aggfunc="std")
```

```python
pd.crosstab(df["Gender"],df["LoanAmount"],values=df["ApplicantIncome"],aggfunc="std")
```

```python
pd.crosstab(df["Gender"],df["LoanAmount"],values=df["ApplicantIncome"],aggfunc="std",normalize=True)
```

```python
pd.crosstab(df["Gender"],df["LoanAmount"],values=df["ApplicantIncome"],aggfunc="count")
```
