## import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## importing dataset

```python
report=pd.read_csv(r"C:\Users\alasy\OneDrive\Desktop\car_sales_data.csv")
report
```

## univariate analysis

```python
report.dtypes
```

```python
report["Car Year"].plot.hist(color="pink")
plt.title("histogram of car sales")
plt.xlabel("sale year")
plt.ylabel("freq/count")
plt.show()
```

```python
plt.hist(report["Commission Rate"],color="cyan")
plt.xlabel("Commission Rate")
plt.ylabel("frequency")
plt.title("histogram of Commission Rate ")
plt.show()
```

```python
sns.distplot(report["Commission Earned"],color="red",kde=True)
plt.xlabel("Commission Earned")
plt.ylabel("count")
plt.title("histogram of Commission Earned")
plt.show()
```

```python
sns.distplot(report["Sale Price"])
plt.show()
```

```python
sns.boxplot(report["Sale Price"],color="blue")
```

```python
report["Commission Rate"].plot.box(vert=False)
plt.show()
```

```python
report["Car Make"].value_counts().plot.bar()
```

```python
sns.countplot(report["Car Year"])
plt.legend()
```

```python
report["Car Year"].value_counts().plot.bar(color=["violet","indigo","blue","green","yellow","orange"])
plt.legend()
```

```python
report["Car Model"].value_counts().plot.line(color="violet")
```

```python
report["Car Year"].value_counts().plot.pie(autopct="%0.2f")
plt.legend()
plt.show()
```

## Bivariate analysis

```python
plt.scatter(report["Commission Rate"],report["Commission Earned"],c="indigo")
plt.xlabel("Commission rate")
plt.ylabel("Commission Earned")
plt.show()
```
