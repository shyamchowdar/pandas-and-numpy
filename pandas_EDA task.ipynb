```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\alasy\Downloads\Iris.csv")
data
```

```python
data.dtypes
```

```python
##  showing data using bar graph with pandas
data["SepalLengthCm"].plot.bar(color="blue")
plt.title("bar on sepal length")
plt.xlabel("sepal length")
plt.ylabel("cm")
plt.show()
```

```python
##showing data using histogram 
data["PetalLengthCm"].plot.hist(color="orange")
plt.title("histogram on petal length")
plt.xlabel("petal length")
plt.ylabel("cm")
plt.show()
```

```python
## showing data using pie chart with matplotlib
plt.pie(data["SepalWidthCm"])
plt.title("pie on sepal width")
plt.xlabel("sepal width")
plt.ylabel("cm")
plt.show()
```

```python
## showing data using histogram with using seaborn 
sns.displot(data["PetalWidthCm"],color="black")
plt.title("Petalwidth")
plt.xlabel("Petalwidth")
plt.ylabel("cm")
plt.show()
```

```python
## box plot
data["SepalLengthCm"].plot.box(vert=False)
```
